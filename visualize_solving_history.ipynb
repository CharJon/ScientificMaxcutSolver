```python
import json
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import os
print(os.getcwd())
```

```python
history = json.load(open("benchmarks_results/13_05/g000981.mc_bound_history.json"))
```

# Number of branch and bound nodes

```python
xrange = np.arange(0,len(np.array(history["bab_nodes"])))
sns.lineplot(x=xrange,
             y=np.array(history["bab_nodes"]))
```

# Dual and primal bounds

```python
sns.lineplot(x=np.array(history["cpu_solving_time"][1:]),
             y=np.array(history["dualbound"][1:]),label="Dual")
sns.lineplot(x=np.array(history["cpu_solving_time"][1:]),
             y=np.array(history["primalbound"][1:]), label="Primal")
plt.title("Primal and Dual Bound")
plt.xlabel("Solving time")
plt.savefig("g000981.svg")
plt.show()
```

# Number of LP-Iterations

```python
sns.lineplot(x=np.array(history["cpu_solving_time"]),
             y=np.array(history["lp_iterations"]))
```
